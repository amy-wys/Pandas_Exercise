### Step 1. Import the necessary libraries

```python
import pandas as pd
import yfinance as yfin
import datetime
```

### Step 2. Create your time range (start and end variables). The start date should be 01/01/2015 and the end should today (whatever your today is).

```python
start = datetime.datetime(2015, 1, 1)
end = datetime.datetime.today()
```

### Step 3. Get an API key for one of the APIs that are supported by Pandas Datareader, preferably for AlphaVantage.

If you do not have an API key for any of the supported APIs, it is easiest to get one for [AlphaVantage](https://www.alphavantage.co/support/#api-key). (Note that the API key is shown directly after the signup. You do *not* receive it via e-mail.)

(For a full list of the APIs that are supported by Pandas Datareader, [see here](https://pydata.github.io/pandas-datareader/readers/index.html). As the APIs are provided by third parties, this list may change.)

### Step 4. Use Pandas Datarader to read the daily time series for the Apple stock (ticker symbol AAPL) between 01/01/2015 and today, assign it to df_apple and print it.

```python
df_apple = yfin.download('AAPL', start, end)
df_apple.head()
```

### Step 5. Add a new column "stock" to the dataframe and add the ticker symbol

```python
df_apple['stock'] = 'AAPL'
df_apple.head()
```

### Step 6. Repeat the two previous steps for a few other stocks, always creating a new dataframe: Tesla, IBM and Microsoft. (Ticker symbols TSLA, IBM and MSFT.)

```python
df_tesla = yfin.download('TSLA', start, end)
df_tesla['stock'] = 'TSLA'
df_tesla.head()
```

```python
df_IBM = yfin.download('IBM', start, end)
df_IBM['stock'] = 'IBM'
df_IBM.head()
```

```python
df_Microsoft = yfin.download('MSFT', start, end)
df_Microsoft['stock'] = 'MSFT'
df_Microsoft.head()
```

### Step 7. Combine the four separate dataFrames into one combined dataFrame df that holds the information for all four stocks

```python
df = pd.concat([df_apple, df_tesla, df_IBM, df_Microsoft])
df
```

### Step 8. Shift the stock column into the index (making it a multi-level index consisting of the ticker symbol and the date).

```python
df = df.set_index('stock', append=True)
df
```

### Step 9. Create a dataFrame called vol, with the volume values.

```python
vol = df.filter(['Volume'])
vol
```

### Step 10. Aggregate the data of volume to weekly.
Hint: Be careful to not sum data from the same week of 2015 and other years.

```python
vol_week = vol.rename_axis(index=['dt', 'stock']) #Rename the index to dt and stock
vol_week = vol_week.reset_index() #Reset index back to range index

#Extract year and week value from dt and create 2 new col
vol_week['year'] = vol_week['dt'].map(lambda x : x.year)
vol_week['week'] = vol_week['dt'].map(lambda x : x.week)

#Delete dt column
del vol_week['dt']

#Set year, week, stock to index
vol_week.set_index(['year', 'week', 'stock'], inplace=True)

#Group the data by stock, year, week
vol_week = vol_week.groupby([pd.Grouper(level='stock'), pd.Grouper(level='year'), pd.Grouper(level='week')]).sum()

#Reset index back to range index (col= stock, year, week, Volume)
vol_week = vol_week.reset_index()

#Reshape df through pivot
vol_week = vol_week.pivot(index=['year', 'week'], columns='stock')

#Drop level
vol_week.columns.droplevel(0)
vol_week.columns = vol_week.columns.droplevel(0)

vol_week
```

### Step 11. Find all the volume traded in the year of 2015

```python
year = vol_week.groupby('year').sum()
year.loc[2015:2015]
```
